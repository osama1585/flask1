Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans-

Flask is a micro web framework for Python, designed to be simple, lightweight, and easy to use. It is often categorized as a "micro" framework because it provides the core functionality needed to build web applications but leaves many other features, such as database abstraction and form handling, to be implemented using third-party extensions or libraries. Flask is widely used for building web applications, RESTful APIs, and other web services.

Here are some of the advantages of Flask:

***Lightweight and Minimalistic:*** Flask is designed to be minimalistic and unopinionated. It doesn't impose any particular way of structuring your application, allowing developers the freedom to make choices about how to organize their code.

***Easy to Learn and Use:*** Flask's simplicity makes it easy for beginners to learn and use. The framework's API is straightforward and well-documented, which helps developers get started quickly.

***Extensible:*** Flask follows the WSGI (Web Server Gateway Interface) specification, which means it can be easily integrated with various third-party extensions and libraries to add functionality as needed. There is a large ecosystem of Flask extensions available for common tasks like database integration, authentication, and more.

***Flexibility:*** Flask doesn't impose a specific database, template engine, or other components on you. You can choose the tools and libraries that best suit your project's requirements.

***Built-in Development Server:*** Flask includes a built-in development server, making it easy to develop and test your application without the need for additional setup.

Q3. What is App routing in Flask? Why do we use app routes?

Ans-

Defining Routes: In Flask, you use the @app.route() decorator to associate a URL with a specific function. This decorator specifies which URL should trigger the associated function when a request is made to that URL.

```python
from flask import Flask
```

```python
# example
app=Flask(__name__)

@app.route('/')
def my():
    return 'this is my web'\

if __name__=='__main__':
    app.run(host="0.0.0.0")
    
```

***Why use app Routes--***

***Organizing Code:*** App routing allows you to organize your code by separating different parts of your application into separate functions or view functions. Each route can be associated with a specific functionality or page of your application.

***URL Mapping:*** It provides a way to map URLs to specific views or functions, making it easy to define the structure of your web application.

***HTTP Methods:*** You can associate different HTTP methods (GET, POST, PUT, DELETE, etc.) with specific routes, allowing you to handle various types of requests differently.

***URL Building:*** Flask provides URL building functions that generate URLs for specific routes, making it easy to create links within your application.

***RESTful APIs:*** If you're building a RESTful API, app routing is essential for defining the endpoints and handling different API requests.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Ans-

```python
from flask import Flask

app = Flask(__name__)

```

```python


@app.route('/welcome')
def welcome():
    return "<h1>Welcome to abc Corporation</h1>"

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

```python


@app.route('/')
def company_info():
    info = """
    <h1>Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999</h1>
    """
    return info

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
